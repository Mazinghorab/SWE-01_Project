# IoT-Integrated Smart Warehouse & Inventory Logistics Portal

- A management system for fulfillment centers to track the physical movement of goods, optimize storage, and automate
supplier procurement.
     - Modules: Warehouse Manager, Floor Staff (Picker/Packer), and Supplier.
     - Key Features: Zonal storage optimization, automated stock reorder triggers  (Observer pattern), batch picking route
logic, and supplier performance analytics.


 - ***Detailed Project Overview*** : This is a next-generation fulfillment management ecosystem designed to digitize the physical
movement of goods within a warehouse. By integrating simulated IoT sensor data (tracking temperature, shelf weight, and
location), the platform optimizes storage through Zonal Algorithms, automates procurement using the Observer Pattern,
and streamlines labor through Batch Picking Logic. The system ensures that the "Supply Chain" is never broken, from the
moment a supplier delivers raw pallets to the final second a package is scanned onto a delivery truck.

*Some Website Examples [Example-01](https://www.linnworks.com/) and [Example-02](https://www.shiphero.com/)* 


# Core Modules

1. ***Warehouse Manager Module*** : High-level oversight of inventory health, storage efficiency, and financial
procurement.

2. ***Floor Staff (Picker/Packer) Module*** : Real-time task management for physical movement, guided by optimized
routing.

3. ***Supplier & Procurement Module*** : Automated B2B interface for order fulfillment, lead-time tracking, and quality
auditing.

# Design Functions (42 Functions)

<div class="alert alert-block alert-info">

1. ***Zonal Storage & Inventory Health*** (11 Functions)
    - *Smart Zonal Storage Optimizer* : An algorithm that suggests storage locations based on "Velocity" (how fast an item sells) and physical dimensions.
    - *Volumetric Capacity Calculator* : Logic that prevents overfilling a specific "Zone" or "Aisle" by calculating the 3D volume of current stock vs. bin capacity.
    - *IoT Weight Sensor Simulation* : Logic that detects "Stock Discrepancy" if the recorded weight on a shelf doesn't match the expected weight of the inventory count
    - *Perishable Expiry Watchdog* : An automated system that flags items nearing their "Best Before" date and suggests "First-Expiry, First-Out" (FEFO) picking.
    - *Environmental Sensitivity Monitor* : Logic that triggers an alert if "Sensitive" items (e.g., electronics or perishables) are stored in zones with high simulated temperature/humidity.
    - *Inventory "Reserved" vs. "Available" Logic* : Managing stock states so that items in customer carts are "Soft-Locked" and cannot be sold to others.
    - *Bulk Breakdown Workflow* : A logic gate for "De-palletizing" large shipments into individual SKU units for shelf storage.
    - *Cross-Docking Logic* : Identifying incoming items that are already backordered and routing them directly to "Packing" instead of "Storage."
    - *Inventory Cycle Count Scheduler* : Automatically flagging specific aisles for manual audit based on time since last check or high discrepancy history.
    - *SKU Transition History* : A non-repudiable log of every location a specific item has occupied within the warehouse.
    - *Hazardous Material (HAZMAT) Guard* : Logic that prevents storing incompatible chemicals or materials in the same physical zone.
</div>


<div class="alert alert-block alert-warning">

2. ***Automated Procurement & Supplier Logic*** (11 Functions)
    - *Automated Stock Reorder Trigger (Observer Pattern)* : A system that monitors stock levels and automatically notifies the Supplier module when a "Safety Stock" threshold is hit.
    - *Supplier Lead-Time Estimator* : Logic that predicts when a shipment will arrive based on the supplier's historical delivery performance.
    - *B2B Purchase Order (PO) Generator* : Automatically creating a structured PO with itemized SKUs, prices, and simulated digital signatures.
    - *Supplier Performance Analytics* : Calculating an "Accuracy Score" for suppliers based on the ratio of items ordered vs. items correctly received.
    - *Tiered Supplier Selection Logic* : Automatically choosing a "Primary" or "Secondary" supplier based on current unit price and delivery speed.
    - *Partial Shipment Reconciliation* : Managing the logic when a supplier sends only 50% of an order, keeping the PO state as "Partially Fulfilled."
    - *Inbound Logistics Gatekeeper* : A state-machine for receiving goods: Expected -> At Dock -> Being Inspected -> Stored.
    - *Supplier Quality Control (QC) Workflow* : A guided inspection logic where floor staff must "Approve" or "Reject" items based on damage or incorrect SKUs.
    - *Automated Invoice Matching* : Logic that cross-references the Supplier's Invoice against the Warehouse's "Received Items" log to detect overcharging.
    - *Backorder Management Engine* : Prioritizing incoming stock for customers who have been waiting the longest.
    - *Supplier Contract Versioning* : Storing and managing different pricing agreements and "Volume Discount" tiers for each supplier.
</div>


<div class="alert alert-block alert-success">

3. ***Picking, Packing & Routing*** (10 Functions)
    - *Batch Picking Route Logic* : An algorithm that combines multiple orders into a single "Pick List" to minimize the physical distance a staff member walks.
    - *Real-time Task Load-Balancer* : Automatically assigning picking tasks to floor staff based on their current location and active workload.
    - *Multi-Order "Sort-to-Light" Simulation* : Logic that guides a packer to place picked items into the correct customer-specific shipping bins.
    - *Parcel Weight/Dimension Validator* : Comparing the final packed box weight against the sum of the item weights to detect packing errors.
    - *Shipping Carrier Integration (Simulated)* : A logic gate that selects the best shipping carrier (e.g., Fast vs. Cheap) based on the customer’s delivery address.
    - *Order Fulfillment State-Machine* : Transitions: Processing -> Picking -> Packing -> Shipped -> Delivered.
    - *Pick-Failure Protocol* : A workflow for when an item is missing from its shelf, triggering an immediate "Inventory Audit" and re-routing the picker.
    - *Packing Material Optimizer* : Suggesting the smallest possible box size for an order to reduce shipping costs and waste.
    - *Label Generation & QR-Linkage* : Creating unique, scannable shipping labels that link the physical box to the digital order record.
    - *Fulfillment Speed Analytics* : Tracking the "Seconds-per-Pick" for floor staff to generate productivity reports for managers.
</div>

4. ***System Administration & Security*** (10 Functions)

    - *Role-Based Access Control (RBAC)* : Restricting "Floor Staff" to picking/packing screens while "Managers" have access to procurement and analytics.
    - *System Audit Trail* : A permanent log of every "Inventory Adjustment" (e.g., manually changing a stock count) with a required reason-code.
    - *Warehouse "Emergency Mode" Logic* : A one-click function to pause all picking tasks and lock the dock doors during a simulated security breach.
    - *Device Heartbeat Monitor* : Tracking the "Online" status of handheld scanners and IoT sensors, flagging inactive hardware.
    - *Data Retention & Archive Logic* : Moving "Completed Order" data to long-term storage after 12 months to maintain database performance.
    - *Multi-Warehouse Data Aggregator* : Allowing a manager to view "Global Stock" across multiple physical fulfillment centers.
    - *Warehouse Capacity Alerts* : An observer implementation that notifies the manager when the total warehouse occupancy exceeds 90%.
    - *User Shift/Time-Clock Manager* : Managing the login/logout times of floor staff for labor cost calculation.
    - *Platform-Wide Sustainability Audit* : Calculating the "Carbon Footprint" of the warehouse based on shipping distances and energy use (simulated).
    - *Database Integrity Manager* : A periodic check to ensure every InventoryItem is correctly associated with a Bin and a ProductCategory.
