### Technical Requirements Required
- **Architecture (Design Patterns)** :
    1. *MVC* Architecture (Model-View-Controller)
    2. Object-Oriented Programming *(OOP)*
    3. Singleton Design Pattern for Database Connection

- **Database Connectivity** : *MySQL* (via phpMyAdmin) with *MySQLi* or *PDO*

- **Security & Auth** : 
    1. Authentication using *Password Hashing*
    2. Role-based *Authorization*
    3. PHP Sessions

- **Data Management** : PHP Implementation of *CRUD operations*

- **Features** : 
    1. *File Uploading*
    2. *AJAX* Calls (in one scenario)

### Documentation & Design (UML)
- **Software Requirement Specification (SRS)** : 
    1. *Functional* (Use Case Scenarios) and *Non-Functional Requirements* (for each, both User and System Requirements)
    2. *Use Case Diagram* and it's *Descriptions*
    3. *Activity Diagrams* / *Swimlane Diagrams*

- **Software Design Document (SDD)** :
    1. *Architectural Design* (a block diagram)
    2. *Class Diagrams*
    3. *Sequence Diagrams*
    4. Entity Relationship Diagrams *(ERD)*
    5. *Object Diagrams*
    6. *Communications/Collaboration* Diagrams
    7. *Package Diagrams*
